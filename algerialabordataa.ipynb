# Missing Values pattern analysis 

to achieve this we will first need to calculate the missing values for specific categories , this will help us determine the amount of NAN values in our datasets and better showcase the pattern for later
We will work on calculating : 
- Missing values for each year
- Missing values by indicator 


```python

year_columns = [col for col in AlgeriaLaborData.columns if col.isdigit()]
missing_by_year = AlgeriaLaborData[year_columns].isnull().sum()
missing_pct_by_year = (missing_by_year / len(AlgeriaLaborData)) * 100
```

```python
missing_by_indicator = AlgeriaLaborData[year_columns].isnull().sum(axis=1)
AlgeriaLaborData['missing_count'] = missing_by_indicator
AlgeriaLaborData['data_availability_pct'] = ((len(year_columns) - missing_by_indicator) / len(year_columns)) * 100
```

### visualising :

```python

fig, ax = plt.subplots(figsize=(8, 6))
ax.plot(year_columns, missing_pct_by_year.values, marker='o', linewidth=2)
ax.set_title('Data Availability by Year\n(Lower is Better)', fontsize=14, fontweight='bold')
ax.set_xlabel('Year')
ax.set_ylabel('Missing Data (%)')
ax.grid(True, alpha=0.3)
ax.tick_params(axis='x', rotation=45)
plt.tight_layout()
plt.show()
```

The chart shows a clear inverse relationship as the lower values indicate better data availability
-There's a structural break around 1991, marking a fundamental shift in data collection practices
-The period 2011-2015 represents the golden age of Algerian labor statistics
-Recent years show concerning deterioration in data quality probably due to specific events 
Early Period (1966-1990): Consistently high missing data rates (90-95%) most likely related to civil war and economic adjestments 
Transition Point: Sharp improvement starting around 1991
Volatile Middle Period (1991-2010): Missing data fluctuates between 45-60%
Modern Era (2011-2015): Best data availability period with rates dropping to 20-30%
Recent Decline (2016-2022): Missing data increases again to 40-80%

```python

sample_indicators = AlgeriaLaborData.head(15)
missing_matrix = sample_indicators[year_columns].isnull().astype(int)
fig, ax = plt.subplots(figsize=(10, 6))
sns.heatmap(missing_matrix, cmap='RdYlBu_r', cbar_kws={'label': 'Missing (1) vs Available (0)'}, ax=ax, xticklabels=5, yticklabels=False)
ax.set_title('Missing Data Pattern (Sample Indicators)', fontsize=14, fontweight='bold')
ax.set_xlabel('Year')
plt.tight_layout()
plt.show()
ax.set_title('Missing Data Pattern (Sample Indicators)', fontsize=14, fontweight='bold')
ax.set_xlabel('Year')
```

```python
fig, ax = plt.subplots(figsize=(10, 6))

years_int = [int(year) for year in year_columns]
windows = []
quality_scores = []
for i in range(0, len(years_int) - 4, 5):
    window_years = [str(year) for year in years_int[i:i+5]]
    window_data = AlgeriaLaborData[window_years]
    quality_score = (1 - window_data.isnull().sum().sum() / (len(window_data) * len(window_years))) * 100
    windows.append(f"{years_int[i]}-{years_int[i+4]}")
    quality_scores.append(quality_score)

ax.set_xlabel('Time Period')
ax.set_ylabel('Data Quality Score (%)')
ax.set_xticks(range(len(windows)))
ax.set_xticklabels(windows, rotation=45)
ax.bar(range(len(windows)), quality_scores, alpha=0.7)
ax.set_title('Data Quality Evolution (5-Year Windows)', fontsize=14, fontweight='bold')
ax.set_xlabel('Time Period')
ax.set_ylabel('Data Quality Score (%)')
ax.set_xticks(range(len(windows)))
ax.set_xticklabels(windows, rotation=45)

plt.tight_layout()
plt.savefig('missing_values_analysis.png', dpi=300, bbox_inches='tight')
plt.show()
```

Analysis :

1966-1985: Extremely poor quality (~5-7%)
1986-1990: Slight improvement to ~7%
1991-1995: Dramatic jump to 45% (institutional transformation)
1996-2010: Steady improvement from 47% to 53%
2011-2015: Peak quality at 62%
2016-2020: Decline to 55%
The most dramatic improvement occurs in the 1991-1995 window
Quality improvements are gradual after the initial jump
The peak in 2011-2015 followed by decline suggests recent institutional challenges
The overall trajectory shows successful institutional building with recent setbacks

The oil price boom of the 2000s and early 2010s allowed more resources to be allocated to bureaucratic and statistical infrastructures.

In economic terms, this aligns with Algeriaâ€™s greater macroeconomic stability and a strong push for labor force modernizationâ€”an attempt to bridge informal and formal sector data.

- Recent Decline: Post-2016
Unfortunately, recent years show a worrisome regression in data quality.

The oil crisis post-2014 led to budget cuts and possibly deprioritized data infrastructure.

The political instability and eventual Hirak Movement (2019) further diverted national focus from bureaucratic continuity.

The COVID-19 pandemic threw a wrench in many countries' statistical systems, and Algeria wasnâ€™t spared.

From an economic lens, declining data quality hinders effective labor policy. Without strong data:

## Data availability timeline 
after analysing the missing values and quality evolution over time we will dive on the availability on this time 
I supported this analysis with some major events which i believe played a role in influencing the data collection in Algeria

```python
fig, axes = plt.subplots(2, 1, figsize=(16, 10))

axes[0].fill_between(year_columns, 100-missing_pct_by_year.values, alpha=0.6, label='Data Available')
axes[0].fill_between(year_columns, 100, 100-missing_pct_by_year.values, alpha=0.3, color='red', label='Data Missing')
axes[0].set_title('Algeria Labor Data Availability Timeline (1966-2022)', fontsize=16, fontweight='bold')
axes[0].set_xlabel('Year')
axes[0].set_ylabel('Data Availability (%)')
axes[0].legend()
axes[0].grid(True, alpha=0.3)
axes[0].tick_params(axis='x', rotation=45)

historical_events = {
    1988: "October Riots\n(Political Crisis)",
    1991: "Economic Reforms\nStructural Adjustment",
    2002: "Civil War Ends\nReconciliation",
    2011: "Arab Spring\nProtests",
    2019: "Hirak Movement\nPolitical Change"
}

for year, event in historical_events.items():
    if str(year) in year_columns:
        axes[0].axvline(x=str(year), color='red', linestyle='--', alpha=0.7)
        axes[0].annotate(event, xy=(str(year), 50), xytext=(10, 20), 
                        textcoords='offset points', fontsize=8, 
                        bbox=dict(boxstyle='round,pad=0.3', facecolor='yellow', alpha=0.7))
        
data_phases = []
phase_labels = []
phase_colors = []

for i, year in enumerate(year_columns):
    availability = 100 - missing_pct_by_year[year]
    if availability < 20:
        phase = 0 
        color = 'red'
    elif availability < 50:
        phase = 1  
        color = 'orange'
    elif availability < 80:
        phase = 2  
        color = 'yellow'
    else:
        phase = 3  
        color = 'green'
    
    data_phases.append(phase)
    phase_colors.append(color)

axes[1].scatter(year_columns, data_phases, c=phase_colors, s=50, alpha=0.8)
axes[1].set_title('Data Collection Phases in Algeria', fontsize=14, fontweight='bold')
axes[1].set_xlabel('Year')
axes[1].set_ylabel('Data Quality Phase')
axes[1].set_yticks([0, 1, 2, 3])
axes[1].set_yticklabels(['Poor\n(<20%)', 'Limited\n(20-50%)', 'Good\n(50-80%)', 'Excellent\n(>80%)'])
axes[1].grid(True, alpha=0.3)
axes[1].tick_params(axis='x', rotation=45)

plt.tight_layout()
plt.savefig('data_availability_timeline.png', dpi=300, bbox_inches='tight')
plt.show()
```

Upper Panel: Comprehensive Timeline

Pink Areas: Represent missing data proportions

Lower Panel: Data Collection Phases
The dots colors represent the following :  
Red Dots (Poor <20%): Concentrated in early period and recent years

Yellow Dots (Good 50-80%): Most of the 2000-2020 period

```python
gender_indicators = AlgeriaLaborData[AlgeriaLaborData['Indicator Name'].str.contains('female|male|women|men', case=False, na=False)]
print(f"Found {len(gender_indicators)} gender-related indicators")

key_gender_indicators = [
    'Employment in agriculture, female (% of female employment)',
    'Employment in agriculture, male (% of male employment)',
    'Employment to population ratio, ages 15-24, female (%)',
    'Employment to population ratio, ages 15-24, male (%)'
]
```

```python
key_gender_indicators = [
    'Employment in agriculture, female (% of female employment)',
    'Employment in agriculture, male (% of male employment)',
    'Employment to population ratio, ages 15-24, female (%)',
    'Employment to population ratio, ages 15-24, male (%)'
]
```

```python
fig, ax = plt.subplots(figsize=(10, 6))

recent_years = [str(year) for year in range(2000, 2023) if str(year) in year_columns]

agr_female = AlgeriaLaborData[AlgeriaLaborData['Indicator Name'].str.contains('Employment in agriculture, female', na=False)]
agr_male = AlgeriaLaborData[AlgeriaLaborData['Indicator Name'].str.contains('Employment in agriculture, male', na=False)]

if not agr_female.empty and not agr_male.empty:
    female_data = agr_female[recent_years].iloc[0].dropna()
    male_data = agr_male[recent_years].iloc[0].dropna()
    
    common_years = sorted(list(set(female_data.index) & set(male_data.index)))
    
    ax.plot(common_years, [female_data[year] for year in common_years],
            marker='o', label='Female', linewidth=2)
    ax.plot(common_years, [male_data[year] for year in common_years],
            marker='s', label='Male', linewidth=2)
    
    ax.set_title('Agricultural Employment by Gender\n(% of respective gender employment)',
                 fontsize=14, fontweight='bold')
    ax.set_xlabel('Year')
    ax.set_ylabel('Employment in Agriculture (%)')
    ax.legend()
    ax.grid(True, alpha=0.3)
    ax.tick_params(axis='x', rotation=45)

plt.tight_layout()
plt.show()
```

Male Line (Orange): Starts at 25% (2000), drops sharply to 12% (2010), then stabilizes at 10-12%
Female Line (Blue): Consistently low at 3-4% throughout entire period
Gap Pattern: Starts at ~20 percentage points, narrows to 7-8 points by 2020

Analysis : 

Males show clear structural transformation out of agriculture
Females show minimal change, remaining consistently excluded
The convergence is primarily due to male employment decline, not female employment increase
Recent years show slight uptick in male agricultural employment

```python
fig, ax = plt.subplots(figsize=(10, 6))

recent_years = [str(year) for year in range(2000, 2023) if str(year) in year_columns]

youth_female = AlgeriaLaborData[AlgeriaLaborData['Indicator Name'].str.contains('Employment to population ratio, ages 15-24, female', na=False)]
youth_male = AlgeriaLaborData[AlgeriaLaborData['Indicator Name'].str.contains('Employment to population ratio, ages 15-24, male', na=False)]

if not youth_female.empty and not youth_male.empty:
    female_youth = youth_female[recent_years].iloc[0].dropna()
    male_youth = youth_male[recent_years].iloc[0].dropna()
    
    common_years_youth = sorted(list(set(female_youth.index) & set(male_youth.index)))
    
    ax.plot(common_years_youth, [female_youth[year] for year in common_years_youth],
            marker='o', label='Female', linewidth=2, color='purple')
    ax.plot(common_years_youth, [male_youth[year] for year in common_years_youth],
            marker='s', label='Male', linewidth=2, color='blue')
    
    ax.set_title('Youth Employment-to-Population Ratio by Gender\n(Ages 15-24)',
                 fontsize=14, fontweight='bold')
    ax.set_xlabel('Year')
    ax.set_ylabel('Employment-to-Population Ratio (%)')
    ax.legend()
    ax.grid(True, alpha=0.3)
    ax.tick_params(axis='x', rotation=45)

plt.tight_layout()
plt.show()
```

```python
# Create standalone plot
fig, ax = plt.subplots(figsize=(10, 6))

# Reuse recent years and common_years
recent_years = [str(year) for year in range(2000, 2023) if str(year) in year_columns]

# Assuming agr_female and agr_male were already defined earlier:
if not agr_female.empty and not agr_male.empty:
    female_data = agr_female[recent_years].iloc[0].dropna()
    male_data = agr_male[recent_years].iloc[0].dropna()
    common_years = sorted(list(set(female_data.index) & set(male_data.index)))
    
    gender_gap = []
    gap_years = []
    
    for year in common_years:
        gap = male_data[year] - female_data[year]
        gender_gap.append(gap)
        gap_years.append(year)
    
    ax.bar(gap_years, gender_gap, alpha=0.7, color='coral')
    ax.set_title('Gender Gap in Agricultural Employment\n(Male % - Female %)',
                 fontsize=14, fontweight='bold')
    ax.set_xlabel('Year')
    ax.set_ylabel('Gender Gap (Percentage Points)')
    ax.axhline(y=0, color='black', linestyle='-', alpha=0.5)
    ax.grid(True, alpha=0.3)
    ax.tick_params(axis='x', rotation=45)

plt.tight_layout()
plt.show()
```

Peak Gap: Reaches 11+ percentage points in 2003-2004
Gradual Improvement: Steady decline to 7-8 percentage points by 2020
Persistence: Despite improvement, gap remains substantial
Stability: Recent years show relatively stable gap around 7.5 percentage points

Pattern Analysis

The improvement is gradual and consistent rather than dramatic
Gap reduction appears to be slowing in recent years
The trend suggests structural barriers that are difficult to overcome
Even the "improved" gap represents significant gender inequality

Gender Gap Analysis - Critical Findings
The Gender Gap in Agricultural Employment shows:
Peak Inequality: 2003-2004 with 11+ percentage point gaps
Modest Improvement: Gradual reduction to 7-8 points by 2020
Structural Persistence: Despite political modernization, deep cultural barriers remain
Policy Implication: Targeted interventions needed for rural women's economic inclusion

```python
plt.figure(figsize=(10, 6))

decades = ['1970s', '1980s', '1990s', '2000s', '2010s', '2020s']
decade_ranges = [
    (1970, 1979), (1980, 1989), (1990, 1999),
    (2000, 2009), (2010, 2019), (2020, 2022)
]

decade_availability = []
for start_year, end_year in decade_ranges:
    decade_years = [str(year) for year in range(start_year, end_year + 1) if str(year) in year_columns]
    if decade_years:
        decade_data = AlgeriaLaborData[decade_years]
        availability = (1 - decade_data.isnull().sum().sum() / (len(decade_data) * len(decade_years))) * 100
        decade_availability.append(availability)
    else:
        decade_availability.append(0)

plt.bar(decades, decade_availability, alpha=0.7, color='skyblue')
plt.title('Data Collection by Decade', fontsize=14, fontweight='bold')
plt.xlabel('Decade')
plt.ylabel('Data Availability (%)')
plt.xticks(rotation=45)
plt.grid(True, axis='y', alpha=0.3)
plt.tight_layout()
plt.savefig('data_collection_by_decade.png', dpi=300, bbox_inches='tight')
plt.show()
```
